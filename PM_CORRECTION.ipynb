```python
# Ensure datetime format and timezone match
ds['timestamp'] = pd.to_datetime(ds['timestamp'], utc=True)
df['timestamp'] = pd.to_datetime(df['timestamp'], utc=True)

# Inner join on timestamp
aligned = pd.merge(ds, df, on='timestamp', how='inner')

print(aligned.shape)  # This should match the number of perfectly aligned hourly records
aligned
```

```python
X = aligned[['pm02_5', 'humidity', 'temperature']]
y = aligned['PM25']
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

# Coefficients
a, b, c = model.coef_
d = model.intercept_

print(f"Correction equation:")
print(f"PM2.5_corrected = {a:.3f} * PM2.5_lowcost + {b:.3f} * RH + {c:.3f} * Temp + {d:.3f}")
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

y_pred = model.predict(X)

print("R²:", r2_score(y, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y, y_pred)))
print("MAE:", mean_absolute_error(y, y_pred))
```

```python
import pandas as pd

# Load your CSV file
df = pd.read_csv("okess_corrected_pm.csv")

# Define AQI breakpoints for PM2.5
breakpoints = [
    (0.0, 12.0, 0, 50),
    (12.1, 35.4, 51, 100),
    (35.5, 55.4, 101, 150),
    (55.5, 150.4, 151, 200),
    (150.5, 250.4, 201, 300),
    (250.5, 350.4, 301, 400),
    (350.5, 500.4, 401, 500),
]

# AQI calculation function
def calculate_aqi_pm25(pm):
    if pd.isna(pm):
        return None
    for c_low, c_high, i_low, i_high in breakpoints:
        if c_low <= pm <= c_high:
            aqi = (i_high - i_low) / (c_high - c_low) * (pm - c_low) + i_low
            return round(aqi, 2)
    return None

# Apply the function to the corrected PM2.5 column
df["AQI_PM2.5"] = df["pm02_5_corrected"].apply(calculate_aqi_pm25)

# Save the result to a new CSV file
df.to_csv("okess_aqi.csv", index=False)

# Show first few rows
print(df[["timestamp", "pm02_5_corrected", "AQI_PM2.5"]].head(50))
```
