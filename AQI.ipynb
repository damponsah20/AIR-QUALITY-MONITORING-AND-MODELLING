```python
import pandas as pd

# Load the dataset
df = pd.read_csv('KASS_corrected.csv')

# Convert the 'by_hour' column to datetime format with correct format
df['timestamp'] = pd.to_datetime(df['by_hour'], format='%d-%m-%y %H:%M', errors='coerce')

# Check if there are any NaT values (errors in parsing)
print(df[df['timestamp'].isna()])  # This will show rows with incorrect format

# Drop rows where timestamp couldn't be converted (if necessary)
df = df.dropna(subset=['timestamp'])

# Extract hour from the timestamp
df['hour'] = df['timestamp'].dt.hour

# Display first few rows
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.colors as mcolors

# Load datasets
locations = {
    "KASS": pd.read_csv('KASS_corrected.csv'),
    "Okess": pd.read_csv('Okess_Correcteed.csv'),
    "Kgee": pd.read_csv('Kgee_corrected.csv'),
    "Adventist": pd.read_csv('Adventist_corrected.csv'),
    "Ibadur": pd.read_csv('Ibadur_corrected.csv'),
    "Sakafia": pd.read_csv('sakafia_corrected.csv')
}

# Convert timestamps and extract hour
for name, df in locations.items():
    df['timestamp'] = pd.to_datetime(df['by_hour'], errors='coerce')
    df['hour'] = df['timestamp'].dt.hour
    locations[name] = df  # Save the cleaned dataframe

# AQI categories and colors
aqi_categories = [
    (0, 50, "Good", "green"),
    (51, 100, "Moderate", "yellow"),
    (101, 150, "Unhealthy for Sensitive Groups", "orange"),
    (151, 200, "Unhealthy", "red"),
    (201, 300, "Very Unhealthy", "purple"),
    (301, 500, "Hazardous", "maroon"),
]

# Define color mapping
cmap = mcolors.LinearSegmentedColormap.from_list(
    "AQI", [color for _, _, _, color in aqi_categories]
)
norm = mcolors.BoundaryNorm(
    [level[0] for level in aqi_categories] + [500], cmap.N
)

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(14, 10), sharex=False, sharey=False)

for ax, (name, df) in zip(axes.flat, locations.items()):
    # Group by hour and calculate mean AQI
    hourly_aqi = df.groupby('hour')['AQI'].mean()

    # Plot with color mapping
    bars = ax.bar(hourly_aqi.index, hourly_aqi.values, color=cmap(norm(hourly_aqi.values)))

    # Titles and labels
    ax.set_title(name)
    ax.set_xticks(range(0, 24, 3))  # Show every 3rd hour for readability
    ax.set_xlabel('Hour of the Day')
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Remove y-axis label, as color represents AQI
    ax.set_ylabel('')

# Adjust layout for color bar space
fig.subplots_adjust(right=0.85)

# Add color bar outside the plots
cbar_ax = fig.add_axes([1.04, 0.2, 0.02, 0.6])  # [left, bottom, width, height]
cbar = fig.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap), cax=cbar_ax)
cbar.set_label('AQI Levels')

# Set tick labels for color bar
cbar.set_ticks([level[0] for level in aqi_categories])
cbar.set_ticklabels([f"{value} ({label})" for value, _, label, _ in aqi_categories])
plt.suptitle('AIR QUALITY INDEX BY HOUR')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load datasets
locations = {
    "KASS": pd.read_csv('KASS_corrected.csv'),
    "Okess": pd.read_csv('Okess_Correcteed.csv'),
    "Kgee": pd.read_csv('Kgee_corrected.csv'),
    "Adventist": pd.read_csv('Adventist_corrected.csv'),
    "Ibadur": pd.read_csv('Ibadur_corrected.csv'),
    "Sakafia": pd.read_csv('sakafia_corrected.csv')
}

# Convert timestamps and extract month and hour
for name, df in locations.items():
    df['timestamp'] = pd.to_datetime(df['by_hour'], errors='coerce')
    df['month'] = df['timestamp'].dt.month
    df['hour'] = df['timestamp'].dt.hour
    locations[name] = df  # Save the cleaned dataframe

# Define colors for locations
colors = sns.color_palette("husl", len(locations))

# Define month labels
month_labels = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Create subplots (3 rows, 4 columns for 12 months)
fig, axes = plt.subplots(3, 4, figsize=(16, 10), sharex=False, sharey=False)

# Loop through each month
for i, ax in enumerate(axes.flat):
    month = i + 1  # Month number (1 to 12)
    
    for (name, df), color in zip(locations.items(), colors):
        hourly_aqi = df[df['month'] == month].groupby('hour')['AQI'].mean()
        ax.plot(hourly_aqi.index, hourly_aqi.values, marker='o', label=name, color=color, linewidth=2)

    # Titles and formatting
    ax.set_title(month_labels[i], fontsize=12, fontweight="bold")
    ax.set_xticks(range(0, 24, 3))  # Show every 3rd hour for readability
    ax.grid(True, linestyle="--", alpha=0.5)

# Set common labels
fig.text(0.5, -0.02, "Hour of the Day", ha="center", fontsize=14)
fig.text(-0.01, 0.5, "Mean AQI", va="center", rotation="vertical", fontsize=14)
plt.suptitle("Hourly Air Quality Index (AQI) Trends for Each Month", fontsize=16, fontweight="bold")

# Add a **single** legend outside the plot
handles = [plt.Line2D([0], [0], color=color, lw=2, marker='o', label=name) for name, color in zip(locations.keys(), colors)]
fig.legend(handles=handles, loc="upper right", bbox_to_anchor=(1.11, 1), ncol=1, title="Locations")

plt.tight_layout()
plt.show()
```
