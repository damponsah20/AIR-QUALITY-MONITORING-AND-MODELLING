```python
print(data['PM1'].isna().sum())  # Check the number of NaN values
```

```python
import pandas as pd
import plotly.express as px


# 1. Drop rows with missing PM1 values
data = data.dropna(subset=['PM1'])

# 2. Alternatively, fill missing PM1 values with the mean

data.loc[:, 'PM1'] = data['PM1'].fillna(data['PM1'].mean())


# Verify that there are no NaN values in 'PM1'
print("Number of NaN values in 'PM1':", data['PM1'].isna().sum())

```

```python

data['PM1'] = pd.to_numeric(data['PM1'], errors='coerce')

# After cleaning, you might want to drop any new NaNs introduced by non-numeric conversion
data = data.dropna(subset=['PM1'])


```

```python

fig = px.scatter_mapbox(
    data,
    lon=data['Longitude'],
    lat=data['Latitude'],
    zoom=8,
    color=data['PM1'],
    size=data['PM1'],
    width=1200,
    height=900,
    title='PM1 DISTRIBUTION'
)

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r": 0, "t": 50, "l": 0, "b": 10})
#fig.savefig('PM1_distribution_map.html')

fig.show()
```

```python
import plotly.io as pio

# Save the figure as an HTML file
pio.write_html(fig, file="PM1_Distribution_Map.html", auto_open=False)
```

```python
# Load the data
data = pd.read_excel('C:/Users/ampon/Desktop/PYTHON/mobile.xlsx')
data
# Extract relevant columns
lats = data['Latitude']
lons = data['Longitude']
pm1 = data['PM1']  # Replace with the column name for PM1

# # Create a scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(lons, lats, c=pm1, cmap='viridis', s=50, alpha=0.7)
plt.colorbar(label='PM1 Levels')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Spatial Distribution of PM1')
plt.grid(True)
plt.show()
```

```python
# Drop rows with NaN values in Latitude, Longitude, or PM1 columns
data_cleaned = data.dropna(subset=['Latitude', 'Longitude', 'PM1'])
```

```python
import folium
from folium.plugins import HeatMap

# Initialize a map
m = folium.Map(location=[data_cleaned['Latitude'].mean(), data_cleaned['Longitude'].mean()], zoom_start=10)

# # Add PM1 data as a heatmap
heat_data = [[row['Latitude'], row['Longitude'], row['PM1']] for index, row in data_cleaned.iterrows()]
HeatMap(heat_data, radius=10).add_to(m)

# # Save or display the map
m.save('pm1_distribution_map.html')
m
```
